```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy.polynomial.polynomial import Polynomial

# Load fantasy history
df = pd.read_csv("../data/historical_stats/fantasy_scores_2000_2024.csv")
df = df[df["FantPos"].isin(["QB", "RB", "WR", "TE"])]
df = df.dropna(subset=["Player", "Age", "Year", "Fantasy_Pts"])
df["Age"] = df["Age"].astype(int)

# Rank within each year/position
df["PosRank"] = df.groupby(["Year", "FantPos"])["Fantasy_Pts"].rank(method="min", ascending=False)

# Tier logic
def get_tier(rank):
    if rank <= 5:
        return "Top 5"
    elif rank <= 12:
        return "Top 12"
    elif rank <= 24:
        return "Top 24"
    elif rank <= 36:
        return "Top 36"
    else:
        return "Other"

df["Tier"] = df["PosRank"].apply(get_tier)

# Average fantasy points by age/position/tier
agg = df[df["Tier"] != "Other"].groupby(["FantPos", "Tier", "Age"])["Fantasy_Pts"].mean().reset_index()

# Fit and display regression for each position + tier
positions = ["QB", "RB", "WR", "TE"]
tiers = ["Top 5", "Top 12", "Top 24", "Top 36"]

for pos in positions:
    for tier in tiers:
        subset = agg[(agg["FantPos"] == pos) & (agg["Tier"] == tier)]
        if subset.shape[0] < 5:
            continue  # not enough data to fit

        x = subset["Age"]
        y = subset["Fantasy_Pts"]

        # Fit 2nd-degree polynomial: y = a + bx + cx^2
        coefs = Polynomial.fit(x, y, 2).convert().coef
        a, b, c = coefs
        print(f"{pos} {tier}: y = {a:.2f} + {b:.2f}x + {c:.2f}x¬≤")

        # Optional: plot actual vs fitted
        x_range = np.linspace(x.min(), x.max(), 100)
        y_fit = a + b * x_range + c * x_range**2

        plt.figure(figsize=(8, 5))
        plt.scatter(x, y, label="Actual", color="gray")
        plt.plot(x_range, y_fit, label="Fitted", color="blue")
        plt.title(f"Aging Curve Fit: {pos} {tier}")
        plt.xlabel("Age")
        plt.ylabel("Avg Fantasy Pts")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths (adjust if needed)
qb_path = "../data/historical_stats/qb_2000_2024.csv"
rb_path = "../data/historical_stats/rb_2000_2024.csv"
wr_path = "../data/historical_stats/wr_2000_2024.csv"

# Load all data
qb_df = pd.read_csv(qb_path)
rb_df = pd.read_csv(rb_path)
wr_df = pd.read_csv(wr_path)

# Helper to compute breakout probability by age
def compute_breakout(df, yard_col, age_col, player_col, threshold, label):
    df = df[df[yard_col].notna() & df[age_col].notna()]
    df[age_col] = df[age_col].astype(int)
    df[yard_col] = df[yard_col].astype(float)

    df["Broke_Out"] = df[yard_col] >= threshold
    df_sorted = df.sort_values(by=[player_col, "Year"])
    first_breakouts = df_sorted[df_sorted["Broke_Out"]].groupby(player_col).first().reset_index()

    age_counts = first_breakouts[age_col].value_counts().sort_index()
    breakout_prob_by_age = (age_counts / len(first_breakouts)).reset_index()
    breakout_prob_by_age.columns = ["Age", "Breakout_Probability"]
    breakout_prob_by_age["Position"] = label
    return breakout_prob_by_age

# Compute for each position
wr_breakouts = compute_breakout(wr_df, "Receiving Yds", "Age", "Player", 1000, "WR")
rb_breakouts = compute_breakout(rb_df, "Rushing Yds", "Age", "Player", 1000, "RB")
qb_breakouts = compute_breakout(qb_df, "Yds", "Age", "Player", 4000, "QB")
te_breakouts = compute_breakout(wr_df[wr_df["Pos"] == "TE"], "Receiving Yds", "Age", "Player", 800, "TE")

# Combine and plot
all_breakouts = pd.concat([qb_breakouts, rb_breakouts, wr_breakouts, te_breakouts])

# Plot
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
positions = ["QB", "RB", "WR", "TE"]
for i, pos in enumerate(positions):
    ax = axes[i // 2, i % 2]
    data = all_breakouts[all_breakouts["Position"] == pos]
    ax.bar(data["Age"], data["Breakout_Probability"], color="skyblue")
    ax.set_title(f"{pos} Breakout Probability by Age")
    ax.set_xlabel("Age")
    ax.set_ylabel("Probability")
    ax.grid(axis="y")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# --- File paths ---
attributes_path = "../data/league_data/player_pool_enriched.csv"
scores_path = "../data/historical_stats/fantasy_scores_2000_2024.csv"
combine_path = "../data/historical_stats/combine_results.csv"
college_path = "../data/historical_stats/college_stats.csv"

# --- Load data ---
df = pd.read_csv(attributes_path)
fantasy_scores = pd.read_csv(scores_path)
combine_df = pd.read_csv(combine_path)
college_df = pd.read_csv(college_path)

# --- Aggregate career fantasy scores ---
career_totals = (
    fantasy_scores.groupby(["Player", "FantPos"], as_index=False)
    .agg(
        Career_Fantasy_Pts=("Fantasy_Pts", "sum"),
        Career_Seasons=("Year", "nunique")
    )
)

# --- Merge fantasy scores ---
df = df.merge(
    career_totals,
    how="left",
    left_on=["Full_Name", "Position"],
    right_on=["Player", "FantPos"]
)

# --- Merge combine results ---
df = df.merge(combine_df, how="left", on=["Full_Name", "Position"])

# --- Parse Height ---
def parse_height(val):
    try:
        if isinstance(val, str) and "-" in val:
            feet, inches = val.strip("'").split("-")
            return int(feet) * 12 + int(inches)
    except:
        return np.nan
    return val

if "Height" in df.columns:
    df["Height"] = df["Height"].apply(parse_height)

# --- Aggregate college stats ---
college_agg = college_df.groupby(["Full_Name", "Position"], as_index=False).mean(numeric_only=True)
df = df.merge(college_agg, how="left", on=["Full_Name", "Position"])

# --- Export or view ---
df.to_csv("../data/league_data/merged_rookie_training_data.csv", index=False)
print("‚úÖ Saved merged dataset to: merged_rookie_training_data.csv")

# OPTIONAL: Display preview if in notebook
from IPython.display import display
display(df.head())
```

```python
import pandas as pd
import xgboost as xgb
import numpy as np

# --- Load merged dataset ---
merged_path = "../data/league_data/merged_rookie_training_data.csv"
df = pd.read_csv(merged_path)

# --- Clean height if needed ---
def parse_height(val):
    try:
        if isinstance(val, str) and "-" in val:
            feet, inches = val.strip("'").split("-")
            return int(feet) * 12 + int(inches)
    except:
        return np.nan
    return val

if "Height" in df.columns:
    df["Height"] = df["Height"].apply(parse_height)

# --- Define usable numeric features ---
features = [
    "Height", "Weight", "Forty", "Vertical", "Shuttle", "Cone", "Broad Jump", "Bench",
    "Receiving_Rec", "Receiving_Yds", "Receiving_Y/R", "Receiving_TD",
    "Rushing_Att", "Rushing_Yds", "Rushing_Y/A", "Rushing_TD",
    "Cmp", "Att", "Yds", "TD", "Int", "Rate"
]

# --- Convert all features to numeric if they exist ---
for col in features:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# --- Modeling ---
all_predictions = []

for pos in ["QB", "RB", "WR", "TE"]:
    pos_df = df[df["Position"] == pos].copy()

    train_df = pos_df[(pos_df["Career_Fantasy_Pts"].notna()) & (pos_df["years_exp"] > 3)]
    predict_df = pos_df[pos_df["years_exp"] <= 3].copy()

    print(f"[{pos}] Training on {len(train_df)} | Predicting {len(predict_df)}")

    if train_df.empty or predict_df.empty:
        continue

    # Filter feature columns for this position
    feature_cols = [col for col in features if col in train_df.columns]

    # Drop rows with NaNs in selected features
    train_df = train_df.dropna(subset=feature_cols + ["Career_Fantasy_Pts"])
    X_train = train_df[feature_cols]
    y_train = train_df["Career_Fantasy_Pts"]

    # Train model
    model = xgb.XGBRegressor(n_estimators=200, max_depth=4, learning_rate=0.05)
    model.fit(X_train, y_train)

    # Prepare prediction data
    X_pred = predict_df[feature_cols].copy()
    X_pred = X_pred.fillna(X_train.mean())  # fill missing with mean of training set

    predictions = model.predict(X_pred)
    predict_df["Proj_Career_XGB"] = predictions

    all_predictions.append(
        predict_df[["Full_Name", "Position", "Proj_Career_XGB", "years_exp"]]
    )

# --- Combine and export ---
rookie_proj = pd.concat(all_predictions, ignore_index=True)
rookie_proj = rookie_proj.groupby(["Full_Name", "Position"], as_index=False).agg({
    "Proj_Career_XGB": "mean",
    "years_exp": "first"
})

rookie_proj.to_csv("../data/league_data/projected_rookies_xgb.csv", index=False)
print("‚úÖ Saved: projected_rookies_xgb.csv")
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
import numpy as np

# === Load data ===
root = "../data"
enriched = pd.read_csv(f"{root}/league_data/player_pool_enriched.csv")
scores = pd.read_csv(f"{root}/historical_stats/fantasy_scores_2000_2024.csv")
combine = pd.read_csv(f"{root}/historical_stats/combine_results.csv")
college = pd.read_csv(f"{root}/historical_stats/college_stats.csv")

# === Merge career fantasy totals ===
career_pts = (
    scores.groupby(["Player", "FantPos"], as_index=False)
    .agg(Career_Fantasy_Pts=("Fantasy_Pts", "sum"), Career_Seasons=("Year", "nunique"))
)

df = enriched.merge(career_pts, how="left", left_on=["Full_Name", "Position"], right_on=["Player", "FantPos"])
df = df.merge(combine, how="left", on="Full_Name")
df = df.merge(college, how="left", on="Full_Name")

# === Fix height format ===
def parse_height(val):
    try:
        if isinstance(val, str) and "-" in val:
            feet, inches = val.strip("'").split("-")
            return int(feet) * 12 + int(inches)
    except:
        return np.nan
    return val

if "Height" in df.columns:
    df["Height"] = df["Height"].apply(parse_height)

# === Feature columns ===
possible_features = [
    "Height", "Weight", "Forty_Yard", "Vertical", "Shuttle", "Three_Cone", "Broad_Jump", "Bench_Reps",
    "Receiving_Rec", "Receiving_Yds", "Receiving_Y/R", "Receiving_TD",
    "Rushing_Att", "Rushing_Yds", "Rushing_Y/A", "Rushing_TD",
    "Cmp", "Att", "Yds", "TD", "Int", "Rate"
]

# Coerce to numeric
for col in possible_features:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# === Train + predict per position ===
all_predictions = []
for pos in ["QB", "RB", "WR", "TE"]:
    pos_df = df[df["Position"] == pos].copy()
    train_df = pos_df[pos_df["Career_Fantasy_Pts"].notna() & (pos_df["years_exp"] > 3)]
    predict_df = pos_df[pos_df["years_exp"] <= 3].copy()

    if train_df.empty or predict_df.empty:
        continue

    X_train = train_df[possible_features].dropna()
    y_train = train_df.loc[X_train.index, "Career_Fantasy_Pts"]

    if X_train.empty:
        continue

    model = xgb.XGBRegressor(n_estimators=100, max_depth=4)
    model.fit(X_train, y_train)

    X_pred = predict_df[possible_features]
    predict_df = predict_df.loc[X_pred.dropna().index]
    X_pred = X_pred.dropna()

    if not X_pred.empty:
        predict_df["Proj_Career_XGB"] = model.predict(X_pred)
        all_predictions.append(predict_df[["Full_Name", "Position", "Proj_Career_XGB"]])

# === Save ===
if all_predictions:
    out = pd.concat(all_predictions, ignore_index=True)
    out = out.groupby(["Full_Name", "Position"], as_index=False).agg({"Proj_Career_XGB": "mean"})
    out.to_csv(f"{root}/league_data/projected_rookies_xgb.csv", index=False)
    print("‚úÖ Saved: projected_rookies_xgb.csv")
else:
    print("‚ö†Ô∏è No valid rookie projections generated.")
```
